```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
numerical_cols=['index', 'subject#', 'age', 'sex', 'test_time', 'motor_UPDRS',
       'total_UPDRS', 'Jitter(%)', 'Jitter(Abs)', 'Jitter:RAP', 'Jitter:PPQ5',
       'Jitter:DDP', 'Shimmer', 'Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'Shimmer:APQ11', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA', 'PPE']
scaler=StandardScaler()
scaled_cols=pd.DataFrame(scaler.fit_transform(df[numerical_cols]),columns=scaler.get_feature_names_out(numerical_cols))
```

```python
scaled_cols
```

```python
x=scaled_cols
y=df['subject#']
```

```python
x.dtypes
```

```python
#SVM
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
svcm=SVC(kernel='linear')
svcm.fit(x_train,y_train)
```

```python
y_pred=svcm.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
```

```python
yd=label_binarize(y, classes=[0,1,2])
```

```python
nc=yd.shape[1]
```

```python
classifier = OneVsRestClassifier(SVC(kernel='linear',probability=True,random_state=42))
```

```python
y_score=classifier.fit(x_train,y_train).decision_function(x_test)
```

```python
fpr=dict()
tpr=dict()
roc_auc=dict()

for i in range(nc):
    fpr[i],tpr[i],_=roc_curve(y_test==i, y_score[:,i])
    roc_auc[i]=auc(fpr[i],tpr[i])
```

```python
plt.figure()
color=['blue','red','green']
for i,color in zip(range(nc),color):
    plt.plot(fpr[i],tpr[i],color=color,lw=2,label='ROC (area={:.2f}) for class{}'.format(roc_auc[i],df['subject#'][i]))
plt.plot([0,1],[0,1],'k--',lw=2)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
plt.legend(loc='lower right')
plt.show()
           
```
